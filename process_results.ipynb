# EHR study results 

```python
elixhauser_phi = pd.read_csv("data/elixhauser_phi.csv", index_col=0)
elixhauser_intercepts = pd.read_csv("data/elixhauser_intercepts.csv", index_col=0)
elixhauser_gradients = pd.read_csv("data/elixhauser_gradients.csv", index_col=0)
elixhauser_cluster_proba = pd.read_csv("data/elixhauser_cluster_proba.csv", index_col=0)
```

```python
elixhauser_phi.loc["Cluster 23, FIRST_ARRYTHMIA", :]
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
n_clusters = int(elixhauser_phi.index[-1].split(",")[0].split(" ")[1])
```

```python
cluster_idx = [
    int(elixhauser_phi.index[i].split(",")[0].split(" ")[1]) <= n_clusters
    for i in range(len(elixhauser_phi.index))
]
```

```python
phi = elixhauser_phi[cluster_idx]
```

```python
phi.loc["Cluster 23, FIRST_ARRYTHMIA", :]
```

```python
phi.loc["Cluster 23, FIRST_CHF", :]
```

```python
phi.loc[:, "cluster_idx"] = [int(phi.index[i].split(",")[0].split(" ")[1]) for i in range(len(phi.index))]
phi.loc[:, "disease"] = [phi.index[i].split(",")[1].strip() for i in range(len(phi.index))]
```

```python
phi_table = pd.pivot_table(
    phi[["mean", "cluster_idx", "disease"]],
    values="mean",
    index=["disease"],
    columns=["cluster_idx"],
    aggfunc="sum"
)
```

```python
phi_table.loc["Cluster size", :] = elixhauser_cluster_proba["mean"].values
phi_table.loc["Outcome", :] = sigmoid(elixhauser_intercepts["mean"]).values
```

```python
mapping = {
    "FIRST_CPD": "COPD",
    "FIRST_CANCER": "Cancer",
    "FIRST_HYPERTENSION": "Hypertension",
    "FIRST_DEPRESSION": "Depression",
    "FIRST_ARRYTHMIA": "Arrythmia",
    "FIRST_HYPOTHYROIDISM": "Hypothyroidism",
    "FIRST_OND": "Other neurological disorder",
    "FIRST_PSYCHOSES": "Psychosis",
    "FIRST_RENAL_DISEASE": "Renal failure",
    "FIRST_LIVER_DISEASE": "Liver disease",
    "FIRST_PARALYSIS": "Paralysis",
    "FIRST_METASTATIC_CANCER": "Metastatic cancer",
    "FIRST_LYMPHOMA": "Lymphoma",
    "FIRST_FED": "Fluid & electrolyte disorders",
    "FIRST_PCD": "Pulmonary ciculation disorders",
    "FIRST_DRUG_ABUSE": "Drug abuse",
    "FIRST_DEF_ANAEMIA": "Deficiency anaemia",
    "FIRST_VALVULAR_DISEASE": "Valvular disease",
    "FIRST_PEP_ULCER": "Peptic ulcer",
    "FIRST_PVD": "Peripheral vascular disease",
    "FIRST_RA": "Rheumatoid arthritis",
    "FIRST_WEIGHT_LOSS": "Weight loss",
    "FIRST_OBESITY": "Obesity",
    "FIRST_DIAB": "Diabetes",
    "FIRST_DIAB_COMP": "Diabetes with complication",
    "FIRST_CHF": "Congestive heart failure",
    "FIRST_HYPERTENSION_COMP": "Hypertension with complication",
    "FIRST_COAGULOPATHY": "Coagulopathy",
    "FIRST_BL_ANAEMIA": "Blood loss anaemia"
}
```

```python
cluster_sizes = phi_table.loc["Cluster size"].to_numpy()
cluster_order = np.argsort(cluster_sizes)[::-1]

idx = cluster_order[0]
cluster_sizes[idx]
```

```python
l_n = 10
fig, ax = plt.subplots(figsize=(l_n, l_n * 11/16))

locs = np.argsort(phi_table.loc["Outcome"].to_numpy())

cluster_sizes = phi_table.loc["Cluster size"].to_numpy()
cluster_order = np.argsort(np.argsort(cluster_sizes[::-1])[::-1])[::-1] 

c = ax.pcolor(
    phi_table
        .rename(index=mapping)
        .loc[:, locs]
)

ax.set_yticks(np.arange(phi_table.shape[0]) + 0.5, minor=False)
ax.set_xticks(np.arange(phi_table.shape[1]) + 0.5, minor=False)

ax.set_yticklabels(phi_table.rename(index=mapping).index)
ax.set_xticklabels(
    cluster_order[locs] + 1
)

fig.colorbar(c, ax=ax, label="Disease probability")
plt.xlabel("Cluster")
plt.tight_layout()
plt.savefig("plots/SAIL_probability_heatmap_linscale.png", dpi=300)
plt.show()
```

```python
def log_odds(x):
    return np.log(x / (1 - x))

l_n = 10
fig, ax = plt.subplots(figsize=(l_n, l_n * 11/16))

locs = np.argsort(phi_table.loc["Outcome"].to_numpy())

cluster_sizes = phi_table.loc["Cluster size"].to_numpy()
cluster_order = np.argsort(np.argsort(cluster_sizes[::-1])[::-1])[::-1] 

z =  log_odds(
    phi_table
        .rename(index=mapping)
        .loc[:, locs]
)
min_val= np.nanmin(z[~np.isinf(z)].to_numpy().reshape(-1))
z[np.isinf(z)] = min_val - 0.1
c = ax.pcolor(
    z,
)

ax.set_yticks(np.arange(phi_table.shape[0]) + 0.5, minor=False)
ax.set_xticks(np.arange(phi_table.shape[1]) + 0.5, minor=False)

ax.set_yticklabels(phi_table.rename(index=mapping).index)
ax.set_xticklabels(
    cluster_order[locs] + 1
)
fig.colorbar(c, ax=ax, label="Disease log-odds")
plt.xlabel("Cluster")
plt.tight_layout()
plt.savefig("plots/SAIL_probability_heatmap_logscale.png", dpi=300)
plt.show()
```

# EHR study - Age stratification

```python
strata = ["20_40","40_60","60_80","80_100"]

for stratum in strata:
    pr_data = pd.read_csv(f"data/profile_regression_results_age_stratum_{stratum}.csv")
    resp_model = pr_data[1:7]
    resp_model = resp_model.loc[:, ["Unnamed: 0", "mean", "hdi_2.5%", "hdi_97.5%"]]
    resp_model["Coefficient estimate (95% CI)"] = [f"{resp_model["mean"].iloc[i]} ({resp_model["hdi_2.5%"].iloc[i]}, {resp_model["hdi_97.5%"].iloc[i]})" for i in range(len(resp_model.index))]
    resp_model["Odds Ratio (95% CI)"] = [f"{np.exp(resp_model["mean"].iloc[i]):.2f} ({np.exp(resp_model["hdi_2.5%"].iloc[i]):.2f}, {np.exp(resp_model["hdi_97.5%"].iloc[i]):.2f})" for i in range(len(resp_model.index))]
    print(resp_model.loc[
        :, ["Unnamed: 0", "Coefficient estimate (95% CI)", "Odds Ratio (95% CI)"]
        ].rename(
            {"Unnamed: 0": f"Age {stratum.replace("_", " - ")} years"}, axis=1
        ).replace({"AGE_AT_START": "Age", "GNDR_CD": "Female sex", 
                "WIMD_QUINTILE_INCEPTION_1": "WIMD Quintile = 1",
                "WIMD_QUINTILE_INCEPTION_2": "WIMD Quintile = 2",
                "WIMD_QUINTILE_INCEPTION_4": "WIMD Quintile = 4",
                "WIMD_QUINTILE_INCEPTION_5": "WIMD Quintile = 5"})
        .to_latex())
```
